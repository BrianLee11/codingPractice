### 질문 01. 
다음 수열의 n번째 항을 구하는 재귀 알고리즘을 작성하시오.

$a_n = 5a_{n-1} + 3,\quad a_1 = 0$

````python
"""
질문 01. 다음 수열의 n번째 항을 구하는 재귀 알고리즘을 작성하시오.
a_{n} = 5 a_{n-1} + 3, a_{1} = 0
"""
def recursion(n):
    if n == 1:
        return 0
    else:
        return 5 * recursion(n-1) + 3
````

````python
# testing
print(recursion(1)) # 0
print(recursion(2)) # 3
print(recursion(3)) # 18
print(recursion(4)) # 93
````

### 답변 06. 
경계 조건: if n > 0 이 빠지면 move(0, ...) 부분이 어떻게 진행되는지 살펴보자.
```
move(0, a, b, c)             # 0개의 원반을 a에서 b로 옮긴다. c는 보조 기둥.
    move (-1, a, c, b)       # -1개의 원반을 a에서 c로 옮긴다. b는 보조 기둥. -1개의 원반의 의미가 불확실하다.
    a에 있는 원반을 b로 옮긴다  # 옮길 원반이 없다. 
    move (-1, c, b, a)       # -1개의 원반을 c에서 b로 옮긴다. a는 보조 기둥. -1개의 원반의 의미가 불확실하다.
```

- `n`이 0이 되었을 때 종료 조건 없이 계속해서 `n-1`을 호출하므로, `n = -1, -2, ...`로 **무한 재귀**가 발생한다.
  
- `-1개의 원반`은 논리적으로 존재하지 않으며, 호출이 종료되지 않는다면 **영원히 존재하지 않는 원반을 다른 기둥으로 옮기려는 시도**가 반복된다.

- 이로 인해 **스택 오버플로우 (Stack Overflow)** 또는 **무한 루프(looping)** 현상이 발생한다.

#### 🔒 결론

경계 조건 `if n > 0:`은 **반드시 필요**하다.
이 조건이 없으면 함수는 종료되지 않고 **재귀 호출이 무한히 이어져 프로그램이 멈추지 않거나 오류를 낳게 된다.**

n-1 > 1 일 경우 selectionSort(A, n-2)이 실행됩니다. 

- selectionSort(A, n)이 최초 1회 호출된다.

- 조건 n > 1을 만족하면 재귀적으로 selectionSort(A, n-1) 호출.

- 그 다음은 selectionSort(A, n-2), selectionSort(A, n-3), ... 진행.

- 마지막으로 selectionSort(A, 1)이 호출되며, 조건 n > 1이 거짓이므로 재귀가 종료된다.

즉, selectionSort(A[], n) 실행, 

selection(A[], n-1) 실행, 

selection(A[], n-2) 실행, 

selection(A[], n-3) 실행, 
...
selection(A[], 2) 실행, 

selection(A[], 1) 실행 -> 종료. 

```
selectionSort(A, n)
→ selectionSort(A, n-1)
→ selectionSort(A, n-2)
→ ...
→ selectionSort(A, 2)
→ selectionSort(A, 1)  ← 종료 지점
```

총 횟수: n - 1 + 1 = n 번

🧮 총 호출 횟수 계산

호출 단계 수: n-1
마지막 종료 조건을 확인하는 selectionSort(A, 1)까지 포함: $$(n-1) + 1 = n$$

## 답변 08.
pg.39에 따르면 `<변수> = A | B | ... | Z`는 `<변수>는 영문 대문자 중 하나`라는 듯이다.

`<숫자>`는 `0`부터 `9`까지의 **한 자리 숫자** 중 하나를 의미한다.

`<T>`는 다음 두 가지 중 하나:

    - 하나의 숫자 `<숫자>`
    
    - 어떤 `<T>` 뒤에 `<숫자>`가 붙은 것 `<T><숫자>`
